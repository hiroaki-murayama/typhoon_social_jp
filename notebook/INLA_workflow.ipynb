```r
DATA_FILE <- "../data/typhoon_mortality_dummy.csv"
NUM_THREADS <- INLA::inla.getOption("num.threads")
```

```r
source("../src/utils.R")
source("../src/inla_pipeline.R")
source("../src/irr_soclag_prefecture_ci.R")
```

```r
# Run: young
out_young <- run_inla_age("young", data_file=DATA_FILE, seed=1L, num_threads=NUM_THREADS)
out_young$figs
```

```r
# Run: elderly
out_eld <- run_inla_age("eld", data_file=DATA_FILE, seed=1L, num_threads=NUM_THREADS)
out_eld$figs
```

```r
# Write prefecture-level lag x social indicator summaries with 95% CI
write_irr_soclag_prefecture_ci(out_young, "young")
write_irr_soclag_prefecture_ci(out_eld, "eld")
list.files("../output", pattern="irr_soclag_prefecture_ci_(young|eld)\\.csv$", full.names=TRUE)
```

```r
# Key outputs written to output/ and output/fig/
list.files("../output", pattern="young|eld|irr_soclag", full.names=TRUE)
list.files("../output/fig", pattern="young|eld|posterior|exdeath_tc", full.names=TRUE)
```
